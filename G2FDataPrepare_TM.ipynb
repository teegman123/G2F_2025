```python
import pandas as pd

# Define the VCF file path
vcf_file = "/content/drive/MyDrive/G2F2025/Training_data/5_Genotype_Data_All_2014_2025_Hybrids.vcf"

with open(vcf_file, 'r') as file:
    lines = [line.strip() for line in file if not line.startswith('##')]

header = lines[0].lstrip('#').split('\t')

data = [line.split('\t') for line in lines[1:]]

vcf_df = pd.DataFrame(data, columns=header)

vcf_df.to_csv("/content/drive/MyDrive/_allstudy/5_Genotype_Data_All_2014_2025_Hybrids.csv", index=False)

print(vcf_df.head())
```
